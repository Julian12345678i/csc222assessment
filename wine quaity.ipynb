```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Load the dataset
df = pd.read_csv("winequality-white.csv")

# Inspect the data
print("Shape of the dataset:", df.shape)
print("\nInfo:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Rename column names for easier access (optional)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Preview cleaned column names
print("\nRenamed columns:", df.columns)

# Example: convert quality to categorical
df["quality"] = df["quality"].astype("category")
```

```python
# Unique values in quality
print("\nWine quality counts:\n", df["quality"].value_counts())

# Group by quality and get mean of each column
print("\nAverage values by wine quality:\n", df.groupby("quality").mean())

# Basic statistics
print(df.describe(include="all"))
```

```python
# 1. Histogram
df["alcohol"].hist(bins=20)
plt.title("Alcohol Content Distribution")
plt.xlabel("Alcohol")
plt.ylabel("Count")
plt.show()

# 2. Line plot
plt.plot(df["density"])
plt.title("Density Trend")
plt.xlabel("Index")
plt.ylabel("Density")
plt.show()

# 3. Bar chart
df["quality"].value_counts().sort_index().plot(kind='bar')
plt.title("Wine Quality Counts")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()

# 4. Boxplot using Matplotlib
plt.boxplot(df["residual_sugar"])
plt.title("Boxplot of Residual Sugar")
plt.ylabel("Residual Sugar")
plt.show()
```

```python
# 5. Seaborn Histogram
sns.histplot(df["ph"], kde=True)
plt.title("pH Distribution")
plt.show()

# 6. Pairplot
sns.pairplot(df[["alcohol", "ph", "density", "quality"]], hue="quality")
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()

# 7. Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 8. Seaborn Boxplot
sns.boxplot(x="quality", y="alcohol", data=df)
plt.title("Alcohol Content by Quality")
plt.show()
```
